```python
import regex
import matplotlib.pyplot as plt
from math import log, exp

with open("compress_tx_log", "r") as f:
    text = f.read()

matches = regex.findall(
    r"compress tx list n_txs=(\d+) encoded_len=(\d+) encode_time=([\d\.]+.s) compress_time=([\d\.]+.s) compress_len=(\d+) expected_len=(\d+) compressed_ratio=([\d\.]+)",
    text
)

print(f"Found {len(matches)} matches")

for match in list(matches)[-10:]:
    n_txs, encoded_len, encode_time, compress_time, compress_len, expected_len, compressed_ratio = match
    print(
        f"{n_txs} {encoded_len} {encode_time} {compress_time} {compress_len} {expected_len} {compressed_ratio}"
    )

def fit_line(x, y):
    """least squares fit of a line to the data"""
    n = len(x)
    if n == 0:
        return None, None
    x_mean = sum(x) / n
    y_mean = sum(y) / n
    numerator = sum((x[i] - x_mean) * (y[i] - y_mean) for i in range(n))
    denominator = sum((x[i] - x_mean) ** 2 for i in range(n))
    if denominator == 0:
        return None, None
    slope = numerator / denominator
    intercept = y_mean - slope * x_mean
    return slope, intercept


encoded_lens = [int(match[1]) for match in matches]
compress_lens = [int(match[4]) for match in matches]
expected_lens = [int(match[5]) for match in matches]

gen = data.split("\n")
gen = [line.split(",") for line in gen if line.strip()]
gen = [[int(x) for x in line] for line in gen]
gen_encoded_lens = [line[0] for line in gen]
gen_compress_lens = [line[1] for line in gen]

all_encoded_lens = encoded_lens + gen_encoded_lens
all_compress_lens = compress_lens + gen_compress_lens
all_encoded_lens_log = [log(x) for x in all_encoded_lens]
all_compress_lens_log = [log(x) for x in all_compress_lens]
slope, intercept = fit_line(all_encoded_lens_log, all_compress_lens_log)
print(f"Slope: {slope}, Intercept: {intercept}")

predicted_compress_lens = [slope * log(x) + intercept for x in all_encoded_lens]
predicted_compress_lens = [exp(x) for x in predicted_compress_lens]

# draw the expected line with slope and intercept
def predict(encoded_len):
    return exp(0.9881615963738659 * log(encoded_len) + -0.19936358812829091)

for i in [1, 10, 100, 1000, 10000, 100000, 1000000]:
    print(f"Encoded Length: {i}, Predicted Compressed Length: {predict(i)}")

fig = plt.figure(figsize=(20, 20))
plt.scatter(encoded_lens, compress_lens, s=1)
plt.scatter(encoded_lens, expected_lens, s=1)
plt.scatter(gen_encoded_lens, gen_compress_lens, s=1)
plt.scatter(all_encoded_lens, predicted_compress_lens, s=1)

plt.axhline(y=375000, color='r', linestyle='-')

plt.xlabel("Encoded Length")
plt.ylabel("Compressed Length")
plt.title("Encoded Length vs Compressed Length")
plt.xscale("log")
plt.yscale("log")


# start from 0
# plt.xlim(left=1)
# plt.ylim(bottom=1)

plt.grid(True, which="both", linestyle="--", linewidth=0.5)
# plt.savefig("encoded_vs_compressed.png", dpi=300)
plt.show()
```
